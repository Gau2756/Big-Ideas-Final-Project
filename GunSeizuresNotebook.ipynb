```python
import pandas as pd
Gun = pd.read_csv("GunSeizures.csv") 
Gun.head()
Neighborhoods = {}
for i in Gun['neighborhood']:
    Neighborhoods[i] = 0
for (i, f) in zip(Gun['neighborhood'], Gun['total_count']):
    Neighborhoods[i] += f
data = pd.DataFrame()
neighborhood = list(Neighborhoods.keys())
numguns = list(Neighborhoods.values())
data['neighborhood'] = neighborhood
data['number of confiscated guns'] = numguns
data.head(51
         )
```
