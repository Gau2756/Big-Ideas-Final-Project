```python
import pandas as pd

# Set pandas to display all rows
pd.set_option('display.max_rows', None)

# Load the arrests data
Arrests = pd.read_csv("Arrests.csv")

Arrests['INCIDENTNEIGHBORHOOD'] = Arrests['INCIDENTNEIGHBORHOOD'].replace("Central Northside", "Central North Side")


# Count arrests per neighborhood
arrests_per_neighborhood = Arrests['INCIDENTNEIGHBORHOOD'].value_counts().reset_index()
arrests_per_neighborhood.columns = ['Neighborhood', 'Number of Arrests']

# Exclude specific non-neighborhood entries
excluded_neighborhoods = [
    "Mt. Oliver Neighborhood", "Mt. Oliver Boro", "Outside State",
    "Outside City", "Outside County", "Golden Triangle/Civic Arena", "Troy Hill-Herrs Island"
]

filtered_arrests = arrests_per_neighborhood[
    ~arrests_per_neighborhood['Neighborhood'].isin(excluded_neighborhoods)
]

# Reverse the Data so that it goes by lowest arrests to highest
filtered_arrests = filtered_arrests.iloc[::-1].reset_index(drop=True)

filtered_arrests


```
